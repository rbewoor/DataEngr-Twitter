```python
# -*- coding: utf-8 -*-
"""
Created on Thu Dec 20 20:31:21 2018

Read the info from the Kafka topic and write to MongoDB
Using consumer_timeout_ms = <Integer value in millisecs to exit if no new message>
see https://kafka-python.readthedocs.io/en/master/apidoc/KafkaConsumer.html
@author: RB
"""
# copied from https://towardsdatascience.com/kafka-python-explained-in-10-lines-of-code-800e3e07dad1

from kafka import KafkaConsumer
from pymongo import MongoClient
from json import loads
from datetime import datetime
#from kafka.structs import OffsetAndMetadata, TopicPartition

countDocsWritten = 0

print('\nStart time of consumer program: ', datetime.now().strftime("%c"))

topicName = 'TwitterTopicMain1'

consumer = KafkaConsumer(
     bootstrap_servers=['localhost:9092'],
     auto_offset_reset='earliest',
#     auto_offset_reset='latest',   # default is latest
     enable_auto_commit=True,
     #auto_commit_interval_ms=2000,
#     group_id='my-group1',
     consumer_timeout_ms=5000,   #consumer_timeout_ms (int) ‚Äì number of milliseconds to block during 
                                 #    message iteration before raising StopIteration (i.e., ending the 
                                 #    iterator). Default block forever [float(‚Äòinf‚Äô)].
     value_deserializer=lambda x: loads(x.decode('utf-8')))
consumer.subscribe(topicName)

client = MongoClient('localhost:27017')
collection = client.TwitterDbMain1.TwitterColMain1
#
# ways to copy a dictionary value      dict2 = dict(dict1)   OR   dict2 = dict1.copy()
#
for message in consumer:
    msgKey = message.key
    message = message.value
    collection.insert_one(message)
    countDocsWritten = countDocsWritten + 1
    if countDocsWritten % 500 == 0:
        print('Written %d th record' %(countDocsWritten))
    if countDocsWritten % 4000 == 0:
        print('KEY={}\nMESSAGE_VALUE={}\nADDED_TO={}'.format(msgKey ,message, collection))

print('\nWritten %d documents to MongoDb' %(countDocsWritten))
print('\Done processing. Exiting consumer program at: ', datetime.now().strftime("%c"))
```

```python
#Started consumer sometime around 20 mins into the 30 min producer time limit. Processed consumption from kafka and writing to
#    mongo really fast. Once it started reading in real time, slowed down to around 500 writes every 10 seconds.
#    This was not due to mongo writing being slow but because the producer was sending messages slower than the consumer was
#    processing it and writing to mongo.
```
